**Q1. Scrape the details of most viewed videos on YouTube from Wikipedia.**

```python
r=requests.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
r.status_code
```

```python
a=r.content
soup=BeautifulSoup(a,'html.parser')
```

```python
#Creating Empty lists
rnk=[]
nm=[]
ar=[]
up=[]
vi=[]
```

```python
# Extracting the required details
Details=soup.find_all('tr')
```

```python
#Filling empty lists
j=1
for i in range(len(Details)):
    if i>=3 and i<=32:
        nm.append(Details[i].text.split('\n')[2].split('"')[1])
        rnk.append(j)
        ar.append(Details[i].text.split('\n')[3])
        up.append(Details[i].text.split('\n')[5])
        vi.append(Details[i].text.split('\n')[4]+" Billion")
        j+=1
```

```python
#checking length of the  scraped data
len(rnk),len(ar),len(up),len(vi),len(nm)
```

```python
# Creating DataFrame
Most_Watched=pd.DataFrame()
Most_Watched['Rank']=rnk
Most_Watched['Name']=nm
Most_Watched['Artist']=ar
Most_Watched['Upload Date']=up
Most_Watched['Views']=vi
```

```python
Most_Watched.set_index('Rank')
```

```python
economy=driver.find_element_by_xpath('/html/body/div[2]/div[1]/div[2]/div[2]/button')
indian=driver.find_element_by_xpath('/html/body/div[2]/div[1]/div[2]/div[2]/div/a[3]')
```

```python
economy.click()
time.sleep(2)
indian.click()
```

```python
#handling popup ad by clicking anywhere
action = webdriver.common.action_chains.ActionChains(driver)
action.move_to_element_with_offset(economy, 5, 5)
action.click()
action.perform()
```

```python
state=driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/ul/li[1]/a')
time.sleep(2)
state.click()
```

```python
#creating Empty lists
Rank=[]
State=[]
GSDP1=[]
GSDP2=[]
Share=[]
GDP=[]
```

```python
name=driver.find_elements_by_xpath('//td[@class="name"]')
gsd1=driver.find_elements_by_xpath('//td[@class="data"]')
gsd2=driver.find_elements_by_xpath('//td[@class="data sorting_1"]')
```

```python
j=1
for i in name:
    if i.text=='India':
        break
    else:
        State.append(i.text)
        Rank.append(j)
        j+=1
```

```python
for i in range(len(gsd1)):
    if i>165:
        break
    elif i%5==0:
        GSDP1.append(gsd1[i].text)
    elif i%5==1:
        Share.append(gsd1[i].text)
    elif i%5==2:
        GDP.append(gsd1[i].text+ " $ Billion")
```

```python
for i in gsd2:
    if len(GSDP2)>32:
        break
    GSDP2.append(i.text)
```

```python
# Creating DataFrame
GDP_Report=pd.DataFrame()
GDP_Report['Rank']=Rank
GDP_Report['State']=State
GDP_Report['GSDP1']=GSDP1[:33]
GDP_Report['GSDP2']=GSDP2
GDP_Report['Share']=Share
GDP_Report['GDP']=GDP
```

**Q10. Details of Datasets from UCI machine learning repositories.**

```python
#connecting to the web driver
driver=webdriver.Chrome('chromedriver.exe')
```

```python
Url = 'https://archive.ics.uci.edu/ml/index.php'
driver.get(Url)
driver.maximize_window()
```

```python
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
```

```python
all_data=driver.find_element_by_xpath('/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b')
```

```python
all_data.click()
```

```python
name=driver.find_elements_by_xpath('//p[@class="normal"]')
```

```python
for i in range(len(name)):
    if i%7==1 and i>1:
        Dataset_name.append(name[i].text)
    elif i%7==2 and i>2:
        Data_type.append(name[i].text)
    elif i%7==3 and i>3:
        Task.append(name[i].text)
    elif i%7==4 and i>4:
        Attribute_type.append(name[i].text)
    elif i%7==5 and i>5:
        No_of_instances.append(name[i].text)
    elif i%7==6 and i>6:
        No_of_attribute.append(name[i].text)
    elif i%7==0 and i>7:
        Year.append(name[i].text)
```

```python
# Creating DataFrame
ML=pd.DataFrame()
ML['Dataset_name']=Dataset_name[:622]
ML['Data_type']=Data_type[:622]
ML['Task']=Task[:622]
ML['Attribute_type']=Attribute_type[:622]
ML['No_of_instances']=No_of_instances[:622]
ML['No_of_attribute']=No_of_attribute[:622]
ML['Year']=Year[:622]
```
