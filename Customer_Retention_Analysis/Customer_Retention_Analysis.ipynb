```python
for i in cust.columns:
    if cust[i].dtype==object:
        plt.title(str(i))
        sns.countplot(x=i,data=cust,palette='inferno_r')
        plt.xticks(rotation=90,size=11)
        plt.xlabel('')
        plt.show()
```

The no. of females are greater than males. 
People having age between 21-50 years purchase more from online shopping.
People from Delhi are shopping the most through online.
People prefer using mobile internet for online shopping.Amazon seems to be the most popular online platform.

People who shop a lot tend to shop from all online platforms but people who shop less tend to use Myntra and Paytm the least.

```python
a=list(cust['From the following, tick any (or all) of the online retailers you have shopped from;                                                                           '])
b=[]
for i in a:
    b.append(i.split(',')[0])
    try:
        b.append(i.split(',')[1])
        b.append(i.split(',')[2])
        b.append(i.split(',')[3])
    except:
        continue
```

```python
plt.figure(figsize=(8,8))
sns.countplot(b,palette='inferno_r')
plt.title('From the following, tick any (or all) of the online retailers you have shopped from')
plt.xticks(rotation=40,size=12)
plt.show()
```

People love to shop more from Amazon and flipkart.

```python
Recommend=list(cust['Which of the Indian online retailer would you recommend to a friend?'])
Recommend_=[]
for i in Recommend:
    Recommend_.append(i.split(',')[0].replace(' ',''))
    try:
        Recommend_.append(i.split(',')[1].replace(' ',''))
        Recommend_.append(i.split(',')[2].replace(' ',''))
        Recommend_.append(i.split(',')[3].replace(' ',''))
    except:
        continue
```

```python
plt.figure(figsize=(8,8))
sns.countplot(Recommend_,palette='inferno')
plt.title('Which of the Indian online retailer would you recommend to a friend?')
plt.xticks(rotation=40,size=12)
plt.show()
```

People recommend Amazon.in the most to their friends.
